```python
#import library
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
#read data
dt = pd.read_excel('data_query-tbl-details.xlsx')
```

```python
list1 = dt[['Methodologies Used']]
print(list1)

dt['x1_len'] = dt['Methodologies Used'].str.count('^ACM|-ACM')
print(dt['x1_len'] )
print(dt['x1_len'].sum())
# 4330+14 = 4344 (ACM)
```

```python
dt['x2_len'] = dt['Methodologies Used'].str.count('^AMS|-AMS')
print(dt['x2_len'] )
print(dt['x2_len'].sum())
# 3254+29 = 3283 (AMS)
```

```python
dt['x3_len'] = dt['Methodologies Used'].str.count('^AM[^S]|-AM[^S]')
print(dt['x3_len'] )
print(dt['x3_len'].sum())
# 520+24 = 544 (AM)
```

```python
list(dt.columns.values)
dt_count = dt[['Sectoral scopes','Methodologies Used']]
dt_count['Methodologies Used'] = dt_count['Methodologies Used'].str[:3]

dt_count['Methodologies Used'].value_counts()
```

```python
#remove AR- and revise AM0
dt_count = dt_count[dt_count['Methodologies Used']!='AR-']
dt_count['Methodologies Used'] = dt_count['Methodologies Used'].replace(['AM0'],['AM'])
dt_count['Methodologies Used'].value_counts()
```

```python
df_value_counts =dt_count.value_counts()

df_value_counts = df_value_counts.reset_index()
df_value_counts.columns = ['Sectoral scopes','Methodologies Used', 'counts']
df_value_counts
```
